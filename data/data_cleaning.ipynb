```python
import pandas as pd
```

```python
df = pd.read_csv('02_Student_Mental_Health_2021-10-10.csv')
df.head()
```

```python
print(list(df.columns))
```

```python
columns_of_interest = ['Age', 'Ethnicity', 'Sex', 'Gender', 'International', 'Disability', 'Year_credits', 'Year_calendar', 'Program', 'Part_time', 'Degree', 'Living', 'Employment', 'Volunteering', 'Plans', 'Hours_sleep', 'More_sleep', 'Hobbies_Time_1', 'Hobbies_Time_2', 'Hobbies_Time_3', 'Hobbies_Time_4', 'Hobbies_Time_5', 'Hobbies_Time_6', 'Hobbies_Time_7', 'Hobbies_Time_8', 'Strenous exercise', 'Moderate exercise', 'Mild exercise', 'Anaerobic_Aerobic', 'Diagnosis', 'Pre_GAD_1', 'Pre_GAD_2', 'Pre_PHQ_1', 'Pre_PHQ_2', 'Pre_PSS_1', 'Pre_PSS_2', 'Pre_PSS_3', 'Pre_PSS_4', 'Pre_PSS_5', 'Pre_PSS_6', 'Pre_PSS_7', 'Pre_PSS_8', 'Pre_PSS_9', 'Pre_PSS_10']
# also added 'Employment', 'Volunteering', 'Plans'
```

```python
df = df[columns_of_interest]
df.head()
```

```python
pre_pss_columns = ['Pre_PSS_1', 'Pre_PSS_2', 'Pre_PSS_3', 'Pre_PSS_4', 'Pre_PSS_5',
                   'Pre_PSS_6', 'Pre_PSS_7', 'Pre_PSS_8', 'Pre_PSS_9', 'Pre_PSS_10']

df['stress'] = df[pre_pss_columns].mean(axis=1)

# Drop the Pre_PSS columns
df = df.drop(pre_pss_columns, axis=1)

df.head()
```

```python
pre_phq_columns = ['Pre_PHQ_1', 'Pre_PHQ_2']

df['depression'] = df[pre_phq_columns].mean(axis=1)

# Drop the Pre_PHQ columns
df = df.drop(pre_phq_columns, axis=1)

df.head()
```

```python
pre_gad_columns = ['Pre_GAD_1', 'Pre_GAD_2']

df['anxiety'] = df[pre_gad_columns].mean(axis=1)

# Drop the Pre_GAD columns
df = df.drop(pre_gad_columns, axis=1)

df.head()
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
