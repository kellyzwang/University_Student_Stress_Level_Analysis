```python
import pandas as pd
```

```python
data = pd.read_csv('02_Student_Mental_Health_2021-10-10.csv', delimiter = ",")
data
```

```python
data= data.dropna(subset=['StartDate'])
data
```

```python
list(data.columns)
```

```python
df_stress_level = data[['Pre_PSS_1', 'Pre_PSS_2', 'Pre_PSS_3', 'Pre_PSS_4',
            'Pre_PSS_5', 'Pre_PSS_6', 'Pre_PSS_7', 'Pre_PSS_8',
            'Pre_PSS_9', 'Pre_PSS_10']]
data['stress_level'] = df_stress_level.sum(axis=1)/10
data
```

```python
df_sleep = data[['CATS_1', 'CATS_2', 'CATS_3', 'CATS_4', 'CATS_5', 
             'CATS_6', 'CATS_7', 'CATS_8', 'CATS_9', 'CATS_10']]
data['sleep_attitudes'] = df_sleep.sum(axis=1)/10
data
```

```python
data.to_csv('cleaned_stress_data.csv', index=False)
```
